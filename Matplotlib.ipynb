```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
y = [98,67,88,95,88]
x = ["Part1","Part2","Part3","Part4","Part5"]

colors = ["Red","blue","green","yellow","pink"]
plt.bar(x,y, color = colors, edgecolor = "black")
plt.xlabel("Parts of Harry Potter", fontsize = 15)
plt.ylabel("Popularity", fontsize = 15)
plt.title("Popularity of Different Parts of Harry Potter", fontsize = 20 )
plt.savefig("bar.png")
plt.show()
```

Stack Plotü§óüóÉÔ∏èüòç

```python
days = [1,2,3,4,5,6,7]

NOP1 = [5,10,30,20,35,50,60]
NOP2 = [40,50,25,60,50,80,85]
NOP3 = [8,15,25,35,60,75,95]

plt.stackplot(days, NOP1, NOP2, NOP3, colors= ["Green","Red","Yellow"], labels = ["Week1","Week2","Week3"])
plt.legend()
plt.show()
```

Legend in MatplotlibüòçüìÅüóÉÔ∏è

```python
x = [1,2,3,4,5]
y = [45,70,35,68,25]
y1 = [40,60,15,60,14]
y2 = [11,30,55,60,19]
y3 = [11,40,45,80,50]

plt.plot(x,y, label = "male", marker = "o")
plt.plot(x,y1, label = "female" , marker = "^")
plt.plot(x,y2, label = "Kids" , marker = "+")
plt.legend()
plt.plot(x,y3)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [45,70,35,68,25]
y1 = [40,60,15,60,14]
y2 = [11,30,55,60,19]
y3 = [11,40,45,80,50]

plt.plot(x,y, label = "male", marker = "o")
plt.plot(x,y1, label = "female" , marker = "^")
plt.plot(x,y2, label = "Kids" , marker = "+")
plt.plot(x,y3, label = "Blanks")
plt.legend(bbox_to_anchor = (0.8,1.1), ncols= 2, labelspacing = 1)
plt.plot(x,y3)
plt.show()
```

Sub PlotüóÉÔ∏èüòçüóÉÔ∏è

```python
x = [1,2,3,4,5]
y = [20,25,50,27,30]
plt.subplot(2,3,2)  #rows, column, chart no
plt.bar(x, y)
plt.suptitle("Employee Data")
plt.show()

x1 = [5,6,7,8,9]
y2 = [67,50,66,56,82]
plt.subplot(2,3,2)
plt.scatter(x1,y2)
plt.show()

x1 = [2,4,6,8,10]
y2 = [57,50,66,56,22]
plt.subplot(2,3,3)
plt.step(x1,y2)
plt.show()
```
