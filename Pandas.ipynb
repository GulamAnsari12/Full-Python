Merge, Join and Concatenate in Pandasüò¥ü§óüòç

```python
data1 = {"Emp ID":["E01","E02","E03","E04","E05","E06"],
         "Name" :["Ram","Shyam","Rahul","Vishal","Ravi","John"],
         "Age":[34,56,23,44,32,36]}

data2 = {"Emp ID":["E01","E07","E03","E04","E08","E06"],
         "Salary":[45000,56000,34000,30000,50000,62000]}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

df1
```

```python
df2
```

```python
print(pd.merge(df1,df2,on = "Emp ID"))
```

```python
print(pd.merge(left = df1,right = df2, on = "Emp ID", how = "left"))
```

```python
data1 = {"Emp ID":["E01","E02","E03","E04","E05","E06"],
         "Name" :["Ram","Shyam","Rahul","Vishal","Ravi","John"],
         "Age":[34,56,23,44,32,36]}

data2 = {"Emp ID":["E07","E08","E09","E10","E11","E12"],
         "Name" :["Bittu","Chintu","Pappu","Chotu","Bunty","Golu"],
         "Age":[34,56,23,44,32,36]}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
```

```python
print(pd.concat([df1,df2]))
```

Campare Data Frame in Pandasüòçü§óüò¥

```python
dict = {"Fruits":["Mango","Apple","Banana","Papaya"],
        "Price":[100,150,50,35],
        "Quantity":[15,10,10,3]}
df1 = pd.DataFrame(dict)
df1
```

```python
df2 = df1.copy()
df2
```

```python
df2.loc[0,"Price"] = 120
df2.loc[1,"Price"] = 175
df2.loc[3,"Price"] = 30
df2.loc[0,"Quantity"] = 12
df2.loc[1,"Quantity"] = 15
df2.loc[3,"Quantity"] = 5

print(df2)
print(df1)
```

```python
print(df1.compare(df2, align_axis = 0))
```

Pivoting and Melting DataFramesü§óüòçüò¥

```python
dict = {"Keys":["K1","K2","K1","K2"],
        "Names":["John","Ben","David","Peter",],
        "Houses":["Red","Blue","Green","Red"],
        "Grade":["3rd","8th","9th","8th"]}

df = pd.DataFrame(dict)
print(df)
print(df.pivot(index = "Keys",columns = "Names",values= ["Houses","Grade"]))
```

```python
dict = {"Names":["John","Ben","David","Peter",],
        "Houses":["Red","Blue","Green","Red"],
        "Grade":["3rd","8th","9th","8th"]}

df = pd.DataFrame(dict)
print(df)
print(pd.melt(df, id_vars=["Names"], value_vars=["Houses","Grade"]))
```
