Bar PlotüóÉÔ∏èüòçüìÅ

```python
data = sns.load_dataset("tips")
a = np.sum(data["tip"])
print(data)
print(a)
```

```python
sns.barplot(data = data , x = "day", y = "tip", hue="sex", palette="spring" , order=["Sun","Thur","Fri","Sat"], errorbar=("ci", 0))
plt.show()
```
